```python
import os
import pycbc
from pycbc.filter.matchedfilter import match
from pycbc.types.frequencyseries import FrequencySeries
import numpy as np
import bilby
import matplotlib.pyplot as plt
from tqdm import tqdm
from concurrent.futures import ProcessPoolExecutor, as_completed
from pathlib import Path
import sys

worker_module_dir = Path.cwd()
if not (worker_module_dir / "figure3_match_worker.py").exists():
    worker_module_dir = Path("Sanity_checks/match_test_result_plot")
if worker_module_dir.exists():
    sys.path.insert(0, str(worker_module_dir.resolve()))
from figure3_match_worker import find_matchvalues_fixed_m, find_matchvalues_fixed_eta

def run_match_pool(worker, input_arguments, npool):
    results = []
    with ProcessPoolExecutor(max_workers=npool) as executor:
        futures = [executor.submit(worker, args) for args in input_arguments]
        for future in tqdm(as_completed(futures), total=len(futures)):
            results.append(future.result())
    return results

size_axis_label = 15
size_tick_label = 13
size_legend_label = 14
```

```python
Gamma = 0.5772156649015329
Pi = np.pi
MTSUN_SI = 4.925491025543576e-06

def findchirp_chirptime(m1, m2, fmin):
    """
    Time taken from f_min to f_lso (last stable orbit). 3.5PN in fourier phase considered.

    Parameters
    ----------
    m1 : `float`
        Mass of the first body in solar masses.
    m2 : `float`
        Mass of the second body in solar masses.
    fmin : `float`
        Lower frequency cutoff.

    Returns
    -------
    chirp_time : float
        Time taken from f_min to f_lso (last stable orbit frequency).
    """

    # variables used to compute chirp time
    m = m1 + m2
    eta = m1 * m2 / m / m
    c0T = c2T = c3T = c4T = c5T = c6T = c6LogT = c7T = 0.0

    c7T = Pi * (
        14809.0 * eta * eta / 378.0 - 75703.0 * eta / 756.0 - 15419335.0 / 127008.0
    )

    c6T = (
        Gamma * 6848.0 / 105.0
        - 10052469856691.0 / 23471078400.0
        + Pi * Pi * 128.0 / 3.0
        + eta * (3147553127.0 / 3048192.0 - Pi * Pi * 451.0 / 12.0)
        - eta * eta * 15211.0 / 1728.0
        + eta * eta * eta * 25565.0 / 1296.0
        + eta * eta * eta * 25565.0 / 1296.0
        + np.log(4.0) * 6848.0 / 105.0
    )
    c6LogT = 6848.0 / 105.0

    c5T = 13.0 * Pi * eta / 3.0 - 7729.0 * Pi / 252.0

    c4T = 3058673.0 / 508032.0 + eta * (5429.0 / 504.0 + eta * 617.0 / 72.0)
    c3T = -32.0 * Pi / 5.0
    c2T = 743.0 / 252.0 + eta * 11.0 / 3.0
    c0T = 5.0 * m * MTSUN_SI / (256.0 * eta)

    # This is the PN parameter v evaluated at the lower freq. cutoff
    xT = np.power(Pi * m * MTSUN_SI * fmin, 1.0 / 3.0)
    x2T = xT * xT
    x3T = xT * x2T
    x4T = x2T * x2T
    x5T = x2T * x3T
    x6T = x3T * x3T
    x7T = x3T * x4T
    x8T = x4T * x4T

    # Computes the chirp time as tC = t(v_low)
    # tC = t(v_low) - t(v_upper) would be more
    # correct, but the difference is negligble.
    return (
        c0T
        * (
            1
            + c2T * x2T
            + c3T * x3T
            + c4T * x4T
            + c5T * x5T
            + (c6T + c6LogT * np.log(xT)) * x6T
            + c7T * x7T
        )
        / x8T
    )
```

```python
time_of_event = 1126259642.413
post_trigger_duration = 1.
duration = 309.0
analysis_start = time_of_event + post_trigger_duration - duration
sampling_frequency = 2048.
f_min = 20.0

injection_parameters = {'mass_1':1.61, 'mass_2':1.18, 'eccentricity':0.1, 'a_1':0., 'a_2':0., 'tilt_1':0., 'tilt_2':0., 'phi_12':0., 'phi_jl':0., 'luminosity_distance':38.87, 'theta_jn':0.4, 'psi':2.659, 'phase':1.3, 'geocent_time':1126259642.413, 'ra':1.375, 'dec':-1.2108,}
# 1.61+1.18, 1.18/1.61

############# TF2Ecck waveform model ################
waveform_arguments1 = dict(waveform_approximant='TaylorF2Ecck',
                          reference_frequency=f_min, minimum_frequency=f_min)

waveform_generator1 = bilby.gw.WaveformGenerator(
    duration=duration, sampling_frequency=sampling_frequency,
    frequency_domain_source_model=bilby.gw.source.lal_eccentric_binary_black_hole_no_spins,
    parameters=injection_parameters, waveform_arguments=waveform_arguments1)

############# TF2Ecc waveform model ################
waveform_arguments3 = dict(waveform_approximant='TaylorF2Ecc',
                          reference_frequency=f_min, minimum_frequency=f_min, pn_phase_order=6, pn_tidal_order=0)

waveform_generator3 = bilby.gw.WaveformGenerator(
    duration=duration, sampling_frequency=sampling_frequency,
    frequency_domain_source_model=bilby.gw.source.lal_eccentric_binary_black_hole_no_spins,
    parameters=injection_parameters, waveform_arguments=waveform_arguments3)

ifos = bilby.gw.detector.InterferometerList(['L1'])
ifos.set_strain_data_from_zero_noise(
    sampling_frequency=sampling_frequency, duration=duration,
    start_time=analysis_start)
```

## Contour plot panels showing mismatch ${\cal M}^*_\%$ contours as a function of $m$ Vs $e_0$ ($q$ fixed)

```python
# Worker imported from figure3_match_worker.py as find_matchvalues_fixed_m.
```

```python
size = 10
q = 0.73
eta = q/(1+q)**2
npool =6
waveform_generatorb_list = [
    waveform_generator3, # TF2Ecc
    ]
waveform_generatora = waveform_generator1 # TF2Ecck

for waveform_generatorb in waveform_generatorb_list:

    if waveform_generatorb.waveform_arguments['waveform_approximant'] == 'TaylorF2Ecch':
        vs = 'EcckVsEcch'
        wide = 3.86 
    elif waveform_generatorb.waveform_arguments['waveform_approximant'] == 'TaylorF2Ecc':
        vs = 'EcckVsEcc'
        wide = 3.81
    elif waveform_generatorb.waveform_arguments['waveform_approximant'] == 'TaylorF2':
        vs = 'EcckVsTF2'
        wide = 4
    else:
        raise ValueError

    filename = f'mismatchvalues_{vs}_fixed_eta.npy'
    plot_filename = f'mismatchvalues_{vs}_fixed_eta.pdf'
    create_new = False

    waveform_argumentsa = dict(waveform_generatora.waveform_arguments)
    waveform_argumentsb = dict(waveform_generatorb.waveform_arguments)
    psd_array = np.array(ifos[0].power_spectral_density_array)
    input_arguments = [
        (i, size, waveform_argumentsa, waveform_argumentsb, q, psd_array)
        for i in range(size)
    ]

    mismatch_array_ab = np.zeros((size, size))
    if not os.path.exists(filename) or create_new:
        for result in run_match_pool(find_matchvalues_fixed_m, input_arguments, npool):
            i, mismatch_array = result
            mismatch_array_ab[i] = mismatch_array
        np.save(filename, mismatch_array_ab)
    else:
        mismatch_array_ab = np.load(filename)

    x = np.linspace(2.0, 6.0, size) # total_mass
    y = np.linspace(0.01, 0.1, size) # e0
    X, Y = np.meshgrid(x, y)
    Z = mismatch_array_ab

    import matplotlib.pyplot as plt
    import numpy as np  # Make sure you import numpy

    # Enable LaTeX rendering
    plt.rcParams.update({
        "text.usetex": True,
        "font.family": "serif",
        "font.serif": ["Computer Modern Roman"],
        "font.size": size_tick_label,
        "axes.labelsize": size_axis_label,
        "axes.titlesize": size_axis_label,
        "xtick.labelsize": size_tick_label,
        "ytick.labelsize": size_tick_label,
        "legend.fontsize": size_legend_label,
        "legend.title_fontsize": size_legend_label,
    })

    # Sample data setup
    fig, ax = plt.subplots(figsize=(12.5, 1.7))

    # Define a smaller number of contour levels
    levels = np.linspace(Z.min(), Z.max(), 15)  # 5 levels

    # Filled contour plot using a dark colormap
    cp = ax.contourf(Y, X, Z, levels=levels, cmap='viridis', alpha=0.95)
    ax.tick_params(axis='both', which='both', labelsize=size_tick_label)
    ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda value, _: f'{value:.2f}'))

    plt.xlabel(r'$e_0$', fontsize=size_axis_label, labelpad=0.0)
    if vs == 'EcckVsTF2':
        plt.ylabel(r'$m$ ($M_\odot$)', fontsize=size_axis_label)
    label = None
    if vs == 'EcckVsEcch':
        label=r'${\cal M}^*_{\%}$'
    plt.ylabel(r'$m$', fontsize=size_axis_label)
    label=r'${\cal M}^*_{\%}$'
        
    cbar = plt.colorbar(cp, label=label, 
                format='%.1f', 
                # show only 5 levels ticks
                boundaries=np.linspace(Z.min(), Z.max(), 5),
                ticks=np.linspace(Z.min(), Z.max(), 5),
                aspect=5, pad=0.01)
    cbar.ax.tick_params(labelsize=size_tick_label)
    cbar.set_label(label, fontsize=size_axis_label)
    # add extra legend
    ax.text(0.012, 4.9, rf'Fixed $\eta$={eta:.4}', color='w', fontsize=size_legend_label, alpha=1.0, zorder=5, bbox=dict(facecolor='gray', alpha=0.3))

    # horizontal line
    ax.axhline(2.79, color='white', linestyle='--', linewidth=1.5, alpha=0.8)
    ax.text(0.015, 3.3, r'$m=2.79$ $M_\odot$', color='w', fontsize=size_legend_label, alpha=1.0, zorder=5)
    
    # Add arrows indicating direction of mismatch increase
    num_ = 0.088
    ax.annotate('', xy=(num_, 5.8), xytext=(num_, 3.5), 
                arrowprops=dict(arrowstyle='->', color='white', lw=2))
    ax.annotate('', xy=(0.094, 3.5), xytext=(num_, 3.5), 
                arrowprops=dict(arrowstyle='->', color='white', lw=2))
    
    ax.text(0.0665, 4.4, r'increasing ${\cal M}^*_{\%}$ directions', color='white', fontsize=size_legend_label, 
            ha='left', va='center', bbox=dict(facecolor='gray', alpha=0.6))
    
    plt.subplots_adjust(left=0.0, bottom=0.0, right=1.0, top=1.0, wspace=0.0, hspace=0.0)
    plt.tight_layout()
    plt.savefig(plot_filename)
```
